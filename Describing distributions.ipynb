```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from scipy import stats
plt.style.use(['seaborn-muted','fivethirtyeight'])
sns.set_context('talk')
```

```python
med_df = pd.read_csv("/Users/dell/Downloads/data (2).csv")
```

```python
med_df.head()
med_df.info()
```

### features to visualize 
- Vitamin D

- Doc visits

- Total charge


## Vitamin D


```python
col = 'VitD_levels'
feature = med_df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize = (12,6))
sns.histplot(feature, ax= ax,stat = 'density', kde = True)
ax.set_title('Vitamin D levels in Patients')
ax.axvline(mean, color='red', lw=3, label =f"Mean: {mean:.2f}")
ax.axvline(median,ls='--', color='green', lw=3, label =f"Median: {median:.2f}")
ax.axvline(plus_one_std, color='black', lw=3, label = f"+1 std: {plus_one_std:.2f}")
ax.axvline(minus_one_std, color='black', lw=3, label = f"-1 std: {minus_one_std:.2f}")
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.legend();
```

#### the graph above shows continous data, since vitamin d levels dont follow a trend and can be any level, they would be considered continous. It also has no skew and there is a normal curve with a symmetrical distribution of data. it also shows mesokurtic distribution, its not pointy enough to be leptokurtic and not flat enough to be platykurtic. 

```python
# i made my decision just by obersving but will also run this to make sure
kurt = stats.kurtosis(med_df['VitD_levels'], fisher = False)
kurt
# i was right Mesokurtic (Kurtosis ~3)
```

```python
# doctor Visits
col = 'Doc_visits'
feature = med_df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize = (12,6))
sns.histplot(feature, ax= ax,stat = 'density', kde = True)
ax.set_title('Average number of doctor visits')
ax.axvline(mean, color='red', lw=3, label =f"Mean: {mean:.2f}")
ax.axvline(median,ls='--', color='green', lw=3, label =f"Median: {median:.2f}")
ax.axvline(plus_one_std, color='black', lw=3, label = f"+1 std: {plus_one_std:.2f}")
ax.axvline(minus_one_std, color='black', lw=3, label = f"-1 std: {minus_one_std:.2f}")
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.legend();
```

This one is a little harder to read visually, however this data is discreet. There is also a pretty even distribution of data and no positive or negative skew (normal). I will calculate kurt below

```python
kurt = stats.kurtosis(med_df['Doc_visits'], fisher = False)
kurt
# its less than 3 so its considered platykurtic
```

```python
col = 'TotalCharge'
feature = med_df[col]
mean = feature.mean()
median = feature.median()
std = feature.std()
plus_one_std = mean + std
minus_one_std = mean - std
fig,ax = plt.subplots(figsize = (12,6))
sns.histplot(feature, ax= ax,stat = 'density', kde = True)
ax.set_title('Total charges')
ax.axvline(mean, color='red', lw=3, label =f"Mean: {mean:.2f}")
ax.axvline(median,ls='--', color='green', lw=3, label =f"Median: {median:.2f}")
ax.axvline(plus_one_std, color='black', lw=3, label = f"+1 std: {plus_one_std:.2f}")
ax.axvline(minus_one_std, color='black', lw=3, label = f"-1 std: {minus_one_std:.2f}")
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.legend();
```

#### The data here is continous data again this data can be anything and doesnt fall into a pattern, there seems to be positive skew ( positive because more money to gain in charges, maybe not so positive for patients. :)) and eyeballing it again its seems like mesokurtic distribution but will run the code again to confirm

```python
kurt = stats.kurtosis(med_df['TotalCharge'], fisher = False)
kurt
```

#### so it was actually greater than 3 and considered Leptokurtic (Kurtosis >3) . Turns out it was more pointy than i thought, which shows eyeballing can sometimes be ok, but its always best to run the code to see if it catches things your eyes cant.
